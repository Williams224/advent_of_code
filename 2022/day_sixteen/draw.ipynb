```python
import networkx as nx
```

```python
G = nx.DiGraph()

with open('/Users/TimothyW/Fun/avent_of_code/2022/day_sixteen/test_input.txt') as f:
    lines = f.readlines()

edge_connections = {}
for v in lines:
    node = v.split(' ')[1]
    cons = v.split(' ')[9:]    
    edge_cons = [e.strip() for e in cons]
    print(edge_cons)
    print(node)
    G.add_node(node)
    edge_connections[node] = edge_cons

for k,v in edge_connections.items():
    for end in v:
        print(k, end)
        G.add_edge(k, end)
```

```python
nx.draw(G)
```
